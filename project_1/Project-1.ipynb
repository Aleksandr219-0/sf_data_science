3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
import numpy as np
import pandas as pd
import re

def get_experience_months(x):
    if pd.isna(x) or x == 'Не указано':
        return np.nan
    
    x = x.split(' , ')[0]
    
    years = 0
    months = 0
    
    m_year = re.search(r'(\d+)\s+(?:год|года|лет)', x)
    m_month = re.search(r'(\d+)\s+(?:месяц|месяца|месяцев)', x)
    
    if m_year:
        years = int(m_year.group(1))
    if m_month:
        months = int(m_month.group(1))
    
    return years * 12 + months

HH_data['Опыт работы (месяц)'] = HH_data['Опыт работы'].apply(get_experience_months)
HH_data = HH_data.drop(columns=['Опыт работы'])
```

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

```python
import pandas as pd
import numpy as np

rates = pd.read_csv(r"C:\Users\aleks\Downloads\ExchangeRates.csv")

rates = rates[['currency', 'date', 'proportion', 'close']]
rates['date'] = pd.to_datetime(rates['date'], format='%d/%m/%y')
rates['rate'] = rates['close'] / rates['proportion']

currency_map = {
    'руб.': 'RUR',
    'бел.руб.': 'BYR',
    'грн.': 'UAH',
    'тг': 'KZT',
    'сум': 'UZS',
    'сом': 'KGS',
    'манаты': 'AZN',
    'лари': 'GEL',
    'кроны': 'CZK',
    'тенге': 'KZT',
    'евро': 'EUR',
    'доллары': 'USD',
    'долларов': 'USD',
    'фунты': 'GBP',
    'форинты': 'HUF',
    'иены': 'JPY',
    'юани': 'CNY'
}

HH_data['валюта'] = HH_data['ЗП'].apply(lambda x: x.split()[1])
HH_data['сумма'] = HH_data['ЗП'].apply(lambda x: float(x.split()[0]))
HH_data['дата'] = pd.to_datetime(HH_data['Обновление резюме'], dayfirst=True).dt.date
rates['date'] = rates['date'].dt.date

def salary_to_rub(row):
    if row['валюта'] == 'руб.':
        return row['сумма']
    currency_code = currency_map.get(row['валюта'])
    if currency_code is None:
        return np.nan
    rate_row = rates[(rates['currency'] == currency_code) & (rates['date'] == row['дата'])]
    if rate_row.empty:
        return np.nan
    return row['сумма'] * rate_row['rate'].values[0]

HH_data['ЗП (руб)'] = HH_data.apply(salary_to_rub, axis=1)
# HH_data = HH_data.drop(columns=['ЗП', 'валюта', 'сумма', 'дата'])
```

```python
import pandas as pd
import numpy as np

rates = pd.read_csv(r"C:\Users\aleks\Downloads\ExchangeRates.csv")
rates = rates[['currency', 'date', 'proportion', 'close']].copy()

rates['date'] = pd.to_datetime(rates['date'], format='%d/%m/%y')
rates['rate'] = rates['close'] / rates['proportion']

currency_map = {
    'руб.': 'RUB',
    'бел.руб.': 'BYN',
    'грн.': 'UAH',
    'тг': 'KZT',
    'сум': 'UZS',
    'сом': 'KGS',
    'манаты': 'AZN',
    'лари': 'GEL',
    'кроны': 'CZK',
    'тенге': 'KZT',
    'евро': 'EUR',
    'доллары': 'USD',
    'долларов': 'USD',
    'фунты': 'GBP',
    'форинты': 'HUF',
    'иены': 'JPY',
    'юани': 'CNY'
}

HH_data['Обновление резюме'] = pd.to_datetime(HH_data['Обновление резюме'], dayfirst=True)
HH_data['date'] = HH_data['Обновление резюме'].dt.date
rates['date'] = rates['date'].dt.date

HH_data['salary'] = HH_data['ЗП'].apply(lambda x: float(x.split()[0]))
HH_data['currency'] = HH_data['ЗП'].apply(lambda x: x.split()[1])

HH_data['currency'] = HH_data['currency'].replace(currency_map)

rates = rates[['currency', 'date', 'close', 'proportion']].copy()
rates.loc[rates['currency'] == 'RUB', 'close'] = 1
rates.loc[rates['currency'] == 'RUB', 'proportion'] = 1

merged = HH_data.merge(rates, how='left', on=['date', 'currency'])

merged['ЗП (руб)'] = merged['salary'] * merged['close'] / merged['proportion']

HH_data = merged.drop(columns=['ЗП', 'salary', 'currency', 'date', 'close', 'proportion'])
```
